### Задание 1
В первой задаче вам предлагается перемножить две квадратные матрицы двумя способами -- без использования пакета ***numpy*** и с ним.

```python
size = (100, 100)

a = np.random.sample(size)
b = np.random.sample(size)

print(f"Ранг матрицы a = {a.ndim}, размерность = {a.shape};")
print(f"Ранг матрицы b = {b.ndim}, размерность = {b.shape};")
```

Без использования numpy:

```python
def multiply(a, b):
  aRows, aCols = len(a), len(a[0])
  bRows, bCols = len(b), len(b[0])

  if aCols != bRows:
    raise Exception("a.columns must be equal to b.rows")

  matrix = [[0 for _ in range(bCols)] for _ in range(aRows)]

  for i in range(aRows):
    for j in range(bCols):
      for k in range(aCols):
        matrix[i][j] += a[i][k] * b[k][j]

  return matrix
```

C использованием numpy:

```python
def np_multiply(a, b):
  return np.dot(a, b)
```

Засекаем время работы **нашей** функции:

```python
%%time
M1 = multiply(a, b)
```

Засекаем время работы функции **numpy.dot**:

```python
%%time
M2 = np_multiply(a, b)
```

Проверяем **корректность**:

```python
np.abs(np.array(M1) - M2).sum()
```

```python
a = list(np.random.sample((1, 3)))[0]
b = list(np.random.sample((1, 3)))[0]

print(a, b)
```

```python
def scalar_product(a, b):
  result = 0

  if len(a) != len(b):
    raise Exception("Different dimensions")

  for i in range(len(a)):
    result += a[i] * b[i]

  return result
```

```python
def np_scalar_product(a, b):
  return np.dot(a, b)
```

Засекаем время работы **нашей** функции:

```python
%time product_1 = scalar_product(a, b)
```

Засекаем время работы функции **numpy.dot**:

```python
%time product_2 = np_scalar_product(a, b)
```

Проверяем **корректность**:

```python
np.abs(product_1 - product_2).sum()
```

### Задание 3
Напишите функцию, которая находим сумму четных диагональныех элементов квадратной матрицы. Если таких элементов нет, то вывести `0`.

```python
def np_diag_2k(a):
   diagonal = np.diag(a)
   even_elements = diagonal[diagonal % 2 == 0]

   if len(even_elements) == 0:
       return 0

   return np.sum(even_elements)
```

### Задание 4
На вход дан двумерный массив $X$. Напишите функцию, которая для каждой строчки $x = (x_1, x_2, \ldots, x_n)$ массива $X$  строит строчку $s = (s_1, s_2, \ldots, s_n)$, где $s_k=\frac{x_1+...+x_k}{k}$, а затем выдаёт массив из построенных строчек. Используйте библиотеку numpy (вам поможет функция cumsum). Выходом функции должен быть двумерный массив той же формы, что и $X$.

```python
def np_sec_av(*A):
  return np.apply_along_axis(np.cumsum, 1, A)
```

```python
A = sps.uniform.rvs(size=10**3)
print(A)

%time S2 = np_sec_av(A)
```

Пусть задан некоторый одномерный массив $X$. Надо построить новый массив, где все элементы с нечетными индексами требуется заменить на число $a$ (если оно не указано, то на 1). Все элементы с четными индексами нужно возвести в куб. Затем записать элементы в обратном порядке относительно их позиций. Массив $X$ при этом должен остаться *без изменений*. В конце требуется слить массив X с преобразованным X и вывести в обратном порядке.  

Пример:  
X = [100, 200, 300, 400, 500] -> [100, a, 300, a, 500] -> [500^3, a, 300^3, a, 100^3] -> склеиваем -> [100, 200, 300, 400, 500, 500^3, a, 300^3, a, 100^3]

```python
from copy import copy

def np_transformation(X, a=1):
   Y = np.where(np.arange(X.size) % 2 == 0, X ** 3, a)[::-1]
   return np.concatenate((X, Y))[::-1]

print(np_transformation(np.array([100, 200, 300, 400, 500])))
```
